```python
millis_len = 100
points_per_bucket = int(ceil(128 * 0.001 * millis_len))
buckets_per_trial = int(ceil(128 / points_per_bucket))
before_avg_interval_millis = 80
points_before_trial = 128 * 0.001 * before_avg_interval_millis
g_accuracies = []
for (start_point, end_point) in [(x, x + points_per_bucket) for x in range(0, points_per_bucket * buckets_per_trial - 1, points_per_bucket)]:
    print "Start point %d, end point %d" % (start_point, end_point)
    g_accuracy = 0
    k = 0
    for leave_out_epoch in range(0, len(data.epochs) - 1, 2):
        k += 1
        train_feat = []
        train_lab = []
        test_feat = []
        test_lab = []
        for idx, (epoch, label) in enumerate(zip(data.epochs, journal.epochType)):
            trials = np.array_split(epoch, 50)
            for trial_num, trial in enumerate(trials):
                if idx != leave_out_epoch and idx != leave_out_epoch + 1:
                    trial_dat = np.array(trial[start_point:end_point, :])
                    if trial_num > 0:
                        trial_before = np.array(trials[trial_num - 1][-1 * points_before_trial:, :])
                        trial_before = np.mean(trial_before, axis=0)
                        trial_dat -= trial_before
                    trial_dat = trial_dat.mean(axis=0)
                    train_feat.append(trial_dat)
                    train_lab.append(label)
                    
                else:
                    trial_dat = np.array(trial[start_point:end_point, :])
                    if trial_num > 0:
                        trial_before = np.array(trials[trial_num-1][-10:, :])
                        trial_before = np.mean(trial_before, axis=0)
                        trial_dat -= trial_before
                    trial_dat = trial_dat.mean(axis=0)
                    test_feat.append(trial_dat)
                    test_lab.append(label)
                    

        train_feat = np.vstack(train_feat)
        test_feat = np.vstack(test_feat)
        train_lab = np.hstack(train_lab)
        test_lab = np.hstack(test_lab)

        cnf = sklearn.linear_model.LogisticRegression()
        train = cnf.fit
        classify = cnf.predict    

        train(train_feat, train_lab)
        acc = 0
        for (feat, lab) in zip(test_feat, test_lab):
            if classify(feat) == lab:
                acc += 1.0
        print (acc / len(test_feat))
        g_accuracy += acc / len(test_feat)

    print "Global accuracy %f" % (g_accuracy/k)
    g_accuracies.append(g_accuracy/k)
```

```python
from scipy.signal import butter, lfilter
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y

filtered = []
for epoch in data.epochs:
    filtered.append(butter_bandpass_filter(epoch, 9, 12, 128))
```

```python
from sklearn import linear_model
millis_len = 100
points_per_bucket = int(ceil(128 * 0.001 * millis_len))
buckets_per_trial = int(ceil(128 / points_per_bucket))
before_avg_interval_millis = 80
points_before_trial = 128 * 0.001 * before_avg_interval_millis
g_accuracies = []
back_stretch = 5
time_bucket_correct = []
time_bucket_incorrect = []


train_feat = []
train_lab = []
trial_num = []
lure_num_correct = []
lure_num_incorrect = []

# collect features for training. also save trial number
for idx, (epoch, label, lure, correct) in enumerate(zip(data.epochs, journal.epochType, journal.epochLure, journal.epochCorrect)):
    trials = np.array_split(epoch, 50)
    for trialNum, trial in enumerate(trials):
        trial_dat = np.array(trial[start_point:end_point, :])
        if trial_num > 0:
            trial_before = np.array(trials[trialNum - 1][-1 * points_before_trial:, :])
            trial_before = np.mean(trial_before, axis=0)
            trial_dat -= trial_before
        trial_dat = trial_dat.mean(axis=0)
        trial_dat = stats.zscore(trial_dat)
        train_feat.append(trial_dat)
        train_lab.append(label)
        trial_num.append(trialNum)
        if (lure[0][trialNum]) and (correct[trialNum]):
            lure_num_correct.append(len(train_lab) - 1)
        elif (lure[0][trialNum]) and (not correct[trialNum]):
            lure_num_incorrect.append(len(train_lab) - 1)

# learn model
cnf = linear_model.LogisticRegression()
cnf.fit(train_feat, train_lab)

correct_lure_diff = [[] for i in  range(0, back_stretch)]
for idx in lure_num_correct:
    if trial_num[idx] > back_stretch:
        for i in range(0, back_stretch):
            diff = cnf.predict_proba(train_feat[idx - i])[0]
            correct_lure_diff[i].append(diff)

incorrect_lure_diff = [[] for i in range(0, back_stretch)]
for idx in lure_num_incorrect:
    if trial_num[idx] > back_stretch:
        for i in range(0, back_stretch):
            diff = cnf.predict_proba(train_feat[idx - i])[0]
            incorrect_lure_diff[i].append(diff)


time_bucket_correct.append(correct_lure_diff)
time_bucket_incorrect.append(incorrect_lure_diff)
```

```python
def lure_performance((data, journal), classifier, (offset_millis, binwidth_millis)):
    from scipy import stats
    import sklearn
    import numpy as np
    from math import ceil

    from sklearn import linear_model
    classifier = linear_model.LogisticRegression()

    offset = int(ceil(128 * 0.001 * offset_millis))
    millis_len = binwidth_millis
    points_per_bucket = int(ceil(128 * 0.001 * millis_len))
    buckets_per_trial = int(ceil(128 / points_per_bucket))
    before_avg_interval_millis = 80
    points_before_trial = 128 * 0.001 * before_avg_interval_millis
    train_feat = []
    train_lab = []
    corr_rej = []
    false_alarm = []

    for idx, (epoch, label, lure, correctResp) in enumerate(zip(data.epochs[:30], journal.epochType, journal.epochLure, journal.epochCorrect)):
        trials = np.array_split(epoch, 50)
        for trial_num, trial in enumerate(trials):
            trial_dat = np.array(trial[offset:(offset + points_per_bucket), :])
            if trial_num > 0:
                trial_before = np.array(trials[trial_num - 1][-1 * points_before_trial:, :])
                trial_before = np.mean(trial_before, axis=0)
                trial_dat -= trial_before
                trial_dat = trial_dat.mean(axis=0)
                train_feat.append(stats.zscore(trial_dat))
                train_lab.append(label)
                if(lure[0][trial_num] and correctResp[trial_num]):
                    corr_rej.append(stats.zscore(trial_dat))
                if(lure[0][trial_num] and not correctResp[trial_num]):
                    false_alarm.append(trial_dat)




    train_feat = np.vstack(train_feat)
    train_lab = np.hstack(train_lab)


    classifier.fit(train_feat, train_lab)
    corr_rej_diff = [classifier.predict_proba(x) for x in corr_rej]
    false_alarm_diff = [classifier.predict_proba(x) for x in false_alarm]
    return(corr_rej_diff, false_alarm_diff)
```
