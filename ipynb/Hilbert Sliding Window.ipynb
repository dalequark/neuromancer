```python
%matplotlib inline
import numpy as np
from parseEEGRaw import *
import matplotlib.pyplot as plt
from scipy import signal
import sklearn
from scipy import stats
print "Done importing"
```

```python
num_participants = 12
```

```python
accuracies = np.load('./gridsearchhilbertsort.npy')
```

```python
for idx in range(0, len(accuracies)): 
    if 'None' in accuracies[idx]:
        accuracies[idx] = [0.5, 0.5]
    else:
        accuracies[idx] = [float(x) for x in accuracies[idx]]
```

```python
binwidths = np.arange(50, 201, 50)
offsets = np.arange(0,801,20)
gridcells = [(offset, size) for offset in offsets for size in bin_sizes]
```

```python
parts = np.array_split(accuracies, num_participants, axis=0)
```

```python
slices = np.array([np.array(np.array_split(part[:,0], len(binwidths)), dtype='float').T for part in parts])
stdslices = np.array([np.array(np.array_split(part[:,1], len(binwidths)), dtype='float').T for part in parts])
```

```python
plt.figure(121)
plt.pcolor(slices[5,:,:],vmin=0.46, vmax=0.57, cmap=cmap)
plt.figure(122)
plt.pcolor(stdslices[5,:,:],vmin = 10**-5, vmax=1, cmap=cmap, norm=LogNorm())
plt.colorbar()
```

```python
fig, axes = plt.subplots(4,3,sharex=True, sharey=True, figsize=(20,20))
axes = list(np.hstack(axes))
cmap = plt.cm.get_cmap("BuPu")
for i in range(0,12):
    plt.subplot(4, 3,i+1)
    plt.pcolor(np.array(np.array_split(parts[i][:,0], len(binwidths)), dtype='float').T,vmin=0.46, vmax=0.57, cmap=cmap)
    plt.title()
    plt.colorbar()
    plt.xticks(np.arange(0,5), bin_sizes)
    plt.yticks(np.arange(0,40,4), offsets[np.arange(0,40,4)])
    plt.xlabel("Bin Width (millis)")
    plt.ylabel("Offset from Stimulus (millis)")
```

```python
from matplotlib.colors import LogNorm
plt.figure(figsize=(20,15))
cmap = plt.cm.get_cmap("BuPu")
cmap_inv = plt.cm.get_cmap("Reds_r")
pvals = stats.ttest_1samp(slices, 0.5, axis=0)[1]
plt.subplot(2,2,1)
plt.pcolor(slices.mean(axis=0),vmin=0.46, vmax=0.57, cmap=cmap)
plt.colorbar()
plt.title("12-Fold Classifier Accuracy Over 11 Participants\n Hilbert Transform",size=17, y=1.02)
plt.xlabel("Bin Width (ms)", size=15)
plt.ylabel("Offset from stimulus (ms)", size=15)
plt.xticks(np.arange(0,len(bin_sizes)),bin_sizes)
plt.yticks(np.arange(0,40), np.arange(0,801,20))
plt.ylim(0,40)

plt.subplot(2,2,2)
plt.title("P Values",size=17, y=1.02)
plt.pcolor(pvals,vmin = 10**-5, vmax=1, cmap=cmap_inv)
plt.xlabel("Bin Width (ms)", size=15)
plt.ylabel("Offset from stimulus (ms)", size=15)
plt.xticks(np.arange(0,len(bin_sizes)),bin_sizes)
plt.yticks(np.arange(0,40), np.arange(0,801,20))
plt.colorbar()
plt.ylim(0,40)
plt.savefig('/Users/Dale/Dropbox/thesislatex/hilbert_gridsearch')
```
