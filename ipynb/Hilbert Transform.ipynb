```python
from scipy.signal import butter, lfilter


def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a


def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y
```

```python
def filter_bands(data):
    
    """     freq_band_names = ['delta', 'theta', 'alpha', 'beta', 'gamma'] """
    freq_bands = [(1,3), (3,7), (7,12), (12,30), (30,59)]
    filtered_data = []
    for band in freq_bands:
        filtered_data.append(butter_bandpass_filter(data, band[0], band[1], fs))
    return filtered_data
```

```python
def nfold((journal, data), make_classifier, (offset_millis, binwidth_millis), preprocess=None):
    """nfold takes a (data, journal) tuple (objects or filenames), a make_classifier method, and 
        an (offset_millis, binwidth_millis) tuple. make_classifier must return
        a (fit, score) tuple"""

     
    from scipy import stats
    import sklearn
    import numpy as np
    from math import ceil
    millis_len = binwidth_millis
    offset = int(ceil(128 * 0.001 * offset_millis)) 
    points_per_bucket = int(ceil(128 * 0.001 * millis_len))
    buckets_per_trial = int(ceil(128 / points_per_bucket))
    l_accuracies = []
    k = 0
    # leave out every successive two epochs
    for leave_out_epoch in range(0, len(data.epochs) - 1, 2):
        k += 1
        train_feat = []
        train_lab = []
        test_feat = []
        test_lab = []
        for idx, (epoch, label) in enumerate(zip(data.epochs, journal.epochType)):
                trials = np.array_split(epoch, 50)
                for trial_num, trial in enumerate(trials):
                        if offset + points_per_bucket > len(trial):
                            return (None, None)
                        if idx != leave_out_epoch and idx != leave_out_epoch + 1:
                            trial_dat = np.array(trial[offset:(offset + points_per_bucket), :])
                            trial_dat = trial_dat.mean(axis=0)
                            train_feat.append(trial_dat)
                            train_lab.append(label)
                            
                        else:
                            trial_dat = np.array(trial[offset:(offset + points_per_bucket), :])
                            trial_dat = trial_dat.mean(axis=0)
                            test_feat.append(trial_dat)
                            test_lab.append(label)

        train_feat = np.vstack(train_feat)
        test_feat = np.vstack(test_feat)
        train_lab = np.hstack(train_lab)
        test_lab = np.hstack(test_lab)

        (fit, score) = make_classifier()
        fit(train_feat, train_lab)
        acc = 0
        l_accuracies.append(score(test_feat, test_lab))
        
    l_accuracies = np.array(l_accuracies)
    return (l_accuracies.mean(), l_accuracies.std())


def nfold_removeprior((journal, data), make_classifier, (offset_millis, binwidth_millis)):
    """nfold takes a (data, journal) tuple (objects or filenames), a make_classifier method, and 
        an (offset_millis, binwidth_millis) tuple. make_classifier must return
        a (fit, score) tuple"""

     
    from scipy import stats
    import sklearn
    import numpy as np
    from math import ceil
    offset = int(ceil(128 * 0.001 * offset_millis)) 
    millis_len = binwidth_millis
    points_per_bucket = int(ceil(128 * 0.001 * millis_len))
    buckets_per_trial = int(ceil(128 / points_per_bucket))
    before_avg_interval_millis = 80
    points_before_trial = 128 * 0.001 * before_avg_interval_millis
    l_accuracies = []
    k = 0
    # leave out every successive two epochs
    for leave_out_epoch in range(0, len(data.epochs) - 1, 2):
        k += 1
        train_feat = []
        train_lab = []
        test_feat = []
        test_lab = []
        for idx, (epoch, label) in enumerate(zip(data.epochs, journal.epochType)):
            trials = np.array_split(epoch, 50)
            for trial_num, trial in enumerate(trials):
                if offset + points_per_bucket > len(trial):
                    continue
                    #return (None, None)
                if idx != leave_out_epoch and idx != leave_out_epoch + 1:
                    trial_dat = np.array(trial[offset:(offset + points_per_bucket), :])
                    if trial_num > 0:
                        trial_before = np.array(trials[trial_num - 1][-1 * points_before_trial:, :])
                        trial_before = np.mean(trial_before, axis=0)
                        trial_dat -= trial_before
                    trial_dat = trial_dat.mean(axis=0)
                    train_feat.append(stats.zscore(trial_dat))
                    train_lab.append(label)
                    
                else:
                    trial_dat = np.array(trial[offset:(offset + points_per_bucket), :])
                    if trial_num > 0:
                        trial_before = np.array(trials[trial_num-1][-10:, :])
                        trial_before = np.mean(trial_before, axis=0)
                        trial_dat -= trial_before
                    trial_dat = trial_dat.mean(axis=0)
                    test_feat.append(stats.zscore(trial_dat))
                    test_lab.append(label)

        train_feat = np.vstack(train_feat)
        test_feat = np.vstack(test_feat)
        train_lab = np.hstack(train_lab)
        train_lab = [1 if 'faces' in l else 0 for l in train_lab]
        test_lab = np.hstack(test_lab)
        test_lab = [1 if 'faces' in l else 0 for l in test_lab]

        (fit, score) = make_classifier()
        fit(train_feat, train_lab)
        acc = 0
        l_accuracies.append(score(test_feat, test_lab))
        
    l_accuracies = np.array(l_accuracies)
    return (l_accuracies.mean(), l_accuracies.std())
```
