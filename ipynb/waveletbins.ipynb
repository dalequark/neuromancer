```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
grid = np.load('./movingtimebins_wavelet.npy')
```

```python
mean_accs = grid[:,0]
```

```python
NUM_PARTICIPANTS = 12
offsets = np.arange(0, 801, 20)
binwidths = np.arange(50, 201, 50)
lows = [0.5]
freqwidths = [59]

instructs = ['%f %f %f %f' % (low, freqwidth, binwidth, offset) for low in lows for freqwidth in freqwidths for binwidth in binwidths for offset in offsets]
```

```python
parts = [np.array(x) for x in np.array_split(mean_accs, NUM_PARTICIPANTS)]
del parts[4]
#mean_accs = np.dstack(parts)[0]
#mean_accs = np.dstack(np.array_split(mean_accs, len(binwidths))).mean(axis=1)
```

```python
parts_grid = [np.array(np.array_split(part, len(binwidths))) for part in parts]
```

```python
from scipy import stats
from matplotlib.colors import LogNorm
```

```python
np.array(parts_grid).shape
```

```python
cmap = plt.cm.get_cmap("BuPu")
cmap_inv = plt.cm.get_cmap("Reds_r")
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
plt.pcolor(np.array(parts_grid).mean(axis=0).T, cmap=cmap, vmin=0.5, vmax=0.54)
plt.colorbar()
plt.title("15-Fold Classifier Accuracy Over 11 Participants\n Wavelet Analysis",size=17, y=1.02)
plt.xlabel("Bin Width (ms)", size=15)
plt.ylabel("Offset from stimulus (ms)", size=15)
plt.xticks(np.arange(0,len(binwidths)),binwidths)
plt.yticks(np.arange(0,40), offsets)
plt.ylim(0,len(offsets))

plt.subplot(1,2,2)
pvals = stats.ttest_1samp(np.array(parts_grid), 0.5, axis=0)[1]
# this is nan for some reason
pvals[2][-1] = 0.5
plt.pcolor(pvals.T, norm=LogNorm(), cmap=cmap_inv, vmin = 10**-5, vmax=1)
plt.colorbar()
plt.title("P Values",size=17, y=1.02)
plt.xlabel("Bin Width (ms)", size=15)
plt.ylabel("Offset from stimulus (ms)", size=15)
plt.xticks(np.arange(0,len(binwidths)),binwidths)
plt.yticks(np.arange(0,40), offsets)
plt.ylim(0,len(offsets))

plt.savefig('/Users/Dale/Dropbox/thesislatex/wavelet_gridsearch', bbox_inches='tight', padding=1)
```
