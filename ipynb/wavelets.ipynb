```python
def get_wav_transform((low,high), data, channelnum = None, numbins = 10, trialnum = None, epochnum = None):
    import mne
    """ Finds the wavelet transform of EEG data from frequencies low to high (Hz). If channelnum is provided,
        returns data only for that channel.  if trialnum or epochnum are given, treats data as data object, else as raw data """
    if hasattr(data, 'epochs'):
        if (epochnum is not None) and (trialnum is not None):
            dataform = np.dstack(data.epochs[epochnum])
            freqs = np.logspace(np.log10(low),np.log10(high), num=numbins, base=10)
            power = mne.time_frequency.single_trial_power(dataform, 128, freqs, verbose=False)
            power = power[0,:,:,:]
            trial_power = np.array(np.array_split(power, 50, axis=2)[trialnum])
            if channelnum is not None:
                return (trial_power[channelnum,:,:], freqs)
            else:
                return (trial_power[:,:,:], freqs)
        else:
            waved_epochs = []
            for epoch in data.epochs:
                freqs = np.logspace(np.log10(low),np.log10(high), num=numbins, base=10)
                power = mne.time_frequency.single_trial_power(np.dstack(epoch), 128, freqs, verbose=False)
                power = power[0,:,:,:]
                
                if channelnum is not None:
                    waved_epochs.append(power[channelnum,:,:])
                else:
                    waved_epochs.append(power)


            return (waved_epochs, freqs)
                                    
    else:
        # if this is a raw data stream... 
            dataform = np.dstack(data)
            freqs = np.logspace(np.log10(low),np.log10(high), num=numbins, base=10)
            power = mne.time_frequency.single_trial_power(dataform, 128, freqs, verbose=False)
            power = power[0,:,:,:]
            if channelnum is not None:
                return (power[channelnum, :, :], freqs)
            else:
                return (power, freqs)
```

```python
def nfold_wavelet((journal, data), make_classifier, (lower_freq, freq_range), num_bins=15,offset_millis=280, binwidth_millis=90):
    offset_millis = 280
    binwidth_millis = 90
    if lower_freq + freq_range > 60:
        return
    from scipy import stats
    import sklearn
    import numpy as np
    from math import ceil
    offset = int(ceil(128 * 0.001 * binwidth_millis)) 
    millis_len = binwidth_millis
    points_per_bucket = int(ceil(128 * 0.001 * millis_len))
    buckets_per_trial = int(ceil(128 / points_per_bucket))
    before_avg_interval_millis = 80
    points_before_trial = 128 * 0.001 * before_avg_interval_millis
    l_accuracies = []
    k = 0
    # leave out every successive two epochs
    for leave_out_epoch in range(0, len(data.epochs) - 1, 2):
        k += 1
        train_feat = []
        train_lab = []
        test_feat = []
        test_lab = []
        for idx, (epoch_raw, label) in enumerate(zip(data.epochs, journal.epochType)):
            epoch = get_wav_transform((lower_freq, lower_freq + freq_range), epoch_raw)[0]
            trials = np.array_split(epoch, 50, axis=2)
            for trial_num, trial in enumerate(trials):
                if idx != leave_out_epoch and idx != leave_out_epoch + 1:
                    trial_dat = np.array(trial[:, :, offset:offset + points_per_bucket])
                    trial_dat = trial_dat.mean(axis=2) # take average over all time periods
                    train_feat.append(np.hstack(trial_dat))
                    train_lab.append(label)

                else:
                    trial_dat = np.array(trial[:, :, offset:offset + points_per_bucket])
                    trial_dat = trial_dat.mean(axis=2) # take average over all time periods
                    test_feat.append(np.hstack(trial_dat))
                    test_lab.append(label)

        train_feat = np.vstack(train_feat)
        test_feat = np.vstack(test_feat)
        train_lab = np.hstack(train_lab)
        test_lab = np.hstack(test_lab)

        (fit, score) = make_classifier()
        fit(train_feat, train_lab)
        acc = 0
        l_accuracies.append(score(test_feat, test_lab))

    l_accuracies = np.array(l_accuracies)
    return (l_accuracies.mean(), l_accuracies.std())

```
